```python
#create dumy variables
train = pd.read_table('http://bit.ly//kaggletrain', sep = ',')
train.head()
```

```python
# Create dummy, prefix to change the name and iloc to keep the new col and drop the original
sex_dummies  = pd.get_dummies(train.Sex, prefix = 'Sex').iloc[:,1:]

# dummies for var wiht more than 1 cat
embarked_dummies  = pd.get_dummies(train.Embarked, prefix = 'Embarked').iloc[:, 1:]
```

```python
# COncat the dummies to the original dataset
train = pd.concat([train, sex_dummies, embarked_dummies], axis = 1)
train.head()
```

```python
# best way to create dummies at one go instead of writing individual codes
#This method automatically drops old cols and replace them by dummies
pd.get_dummies(train, columns = ['Sex', 'Embarked'] )
```

## Customer Function to identify and remove outliers

```python
def remove_outliers(df):
    outliers = {}
    for col in df.columns:
        if str(df[col].dtype)!= 'object':
            df = df[np.abs(df[col] - df[col].mean())< (3*df[col].std())]
            olrs = df[~np.abs(df[col] - df[col].mean())< (3*df[col].std())]
            outliers = pd.Dataframe(olrs)
    return df, outliers
```
