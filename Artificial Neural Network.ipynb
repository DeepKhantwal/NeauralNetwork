```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('C:/Users/ksa/Documents/Machine Learning A-Z/Machine Learning A-Z/Machine Learning A-Z New/Part 8 - Deep Learning/Section 39 - Artificial Neural Networks (ANN)/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
X = dataset.drop('Exited', axis = 1)
Y = dataset[['Exited']]
X.head()
```

```python
X_1 = pd.get_dummies(X, columns = ['Geography', 'Gender'])
X_1.drop(['Surname', 'RowNumber'],  axis = 1, inplace = True)
X_1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X_1,Y, test_size =.30, random_state = 0)
```

```python
X_test.head()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

## Run Logistic Regression to compare with ANN

```python
from sklearn.linear_model import LogisticRegression
Logreg = LogisticRegression(random_state = 0)
Logreg.fit(X_train, Y_train)
predictions = Logreg.predict(X_test)
```

```python
predictions
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, predictions)
```

## Artificial Neural Network

```python
import keras
# Initialize Neural network
from keras.models import Sequential
# Create layers in ANN
from keras.layers import Dense
```

```python
#intialize Neural Network
classifier = Sequential()
```

```python
# Adding Input and first hidden layer
# Output_dim = hidden layer (random), init = weight assignment, activation = activation function, input_dim = input nodes/featues 
classifier.add(Dense(output_dim = 8, init = 'uniform', activation = 'relu', input_dim = 14))

#Adding the second Hidden Layer
classifier.add(Dense(output_dim = 8, init = 'uniform', activation = 'relu'))

#Adding the Output Layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

#Add Compiler (Stochastic Gradient Descent - 'ADAM')
classifier.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = ['accuracy'] )
```

```python
# Fit the ANN model to training set
classifier.fit(X_train, Y_train, batch_size = 10, nb_epoch = 50)
```

```python
# Evaluate the ANN model on test
ANN_Predict = classifier.predict(X_test)

ANN_Predict
```

```python
# Change the probabilities to binary True Or False to run the model accuracy
ANN_Predict = (ANN_Predict> 0.5)
ANN_Predict
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, ANN_Predict)
```
