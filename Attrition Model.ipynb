## Step2 - Missing Value check and Treatment

## Step 4 - Creatin Dummy variables for String cols with Categories

```python
# Full Dummies of col's with small number of categories
attrition_4 = pd.get_dummies(attrition_3, columns = ['C1', 'C4', 'C5','C9','C10','C12','C13', 'Level'])
```

```python
# Creating Dumies for variables with large number of categories

# To avoid creating dummies for variables with high number of categories
# Lets make a list with most frequent categories of the variable
#Creating function for all set of catergorical variables
top_5 = [x for x in attrition_3.C6.value_counts().sort_values(ascending = False).head(5).index]
top_4 = [x for x in attrition_3.C7.value_counts().sort_values(ascending = False).head(4).index]

def one_hot_top_x(df, df_n, variable, top_x_labels):
    #function to creating dummy variables for most frequent cat/labels
    #we can vary the number to be selected for most frequent cat/labels
    
    for label in top_x_labels:
        df_n[variable+'_'+label] = np.where(df[variable]==label,1,0) 
        
```

```python
# encode/create dummy of variable into top X categories
one_hot_top_x(attrition_3,attrition_4, 'C6', top_5 )
one_hot_top_x(attrition_3,attrition_4, 'C7', top_4 )
attrition_4.head()
```

```python
# Creating 'OTHER CATEGORY' variables with large categories
attrition_4['C6_Others'] = np.where((attrition_4.C6_c + attrition_4.C6_q + attrition_4.C6_w + attrition_4.C6_i + attrition_4.C6_aa) == 0, 1, 0)
attrition_4['C7_Others'] = np.where((attrition_4.C7_v + attrition_4.C7_h + attrition_4.C7_bb + attrition_4.C7_ff) == 0, 1, 0)
```

# Step - 5 Remove Multicollinearity, train test split and Model Building

```python
# Seperating Independant and dependant variables for modeling purpose
attrition_X = attrition_4.drop('Attrited', axis = 1)
attrition_Y = attrition_4['Attrited']
```

```python
# MULTICOLLINEARITY AND DROPPING OF VARIABLES WITH HIGH VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(X, thresh='p'):
        # Taken from https://stats.stackexchange.com/a/253620/53565 and modified
        dropped=True
        while dropped:
            variables = X.columns
            dropped = False
            vif = [variance_inflation_factor(X[variables].values, X.columns.get_loc(var)) for var in X.columns]
            
            max_vif = max(vif)
            if max_vif > thresh:
                maxloc = vif.index(max_vif)
                print(f'Dropping {X.columns[maxloc]} with vif={max_vif}')
                X = X.drop([X.columns.tolist()[maxloc]], axis=1)
                dropped=True
        return X
```

```python
#Running the Custom VIF Function
calculate_vif(attrition_X, thresh=10.0)
```

```python
# Train and test split
# Random State = 0 means you get the same sample everytime you run this
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(attrition_X, attrition_Y, test_size = 0.3, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Logistic Regression and Model Evaluation

```python
from sklearn.linear_model import LogisticRegression

# Step 1 - Making an instance logmodel 
logmodel = LogisticRegression(random_state = 0)

#Fit the instance on the train datasethttp://localhost:8888/notebooks/Attrition%20Model.ipynb#Logistic-Regression-and-Model-Evaluation
logmodel.fit(X_train, Y_train)

# Applying the fitted model on test
predictions = logmodel.predict(X_test)
```

```python
#Model classitication report
from sklearn.metrics import classification_report
classification_report(Y_test, predictions)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, predictions)
```

```python
# coefficeints of the trained model
print('\nCoefficient of model :', logmodel.coef_)

# intercept of the model
print('\nIntercept of model',logmodel.intercept_)
```

```python
# K fold Cross Validation
from sklearn.model_selection import cross_val_score
cross_val_score(logmodel, X_train, Y_train, cv = 5, scoring = 'accuracy').mean()
```
